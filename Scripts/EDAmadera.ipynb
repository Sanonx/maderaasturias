```python
df1 = pd.read_csv(PATH_DIR + 'especiesasturias.csv', encoding='latin1', delimiter=';', skiprows=1)
df1.index = df1['Periodo']
df1.drop(['Periodo'], axis=1, inplace=True)

df2 = pd.read_csv(PATH_DIR + 'especiesconcejo.csv', encoding='latin1', delimiter=';', skiprows=1, index_col=0)
concejossinseparar = df2.index.tolist()
columnas = df2.columns.tolist()
columnas

dfz = pd.DataFrame(data=df2.values, columns=columnas)
dfz.index = (concejossinseparar)
df2 = dfz
df2 = df2.T
concejos = df2.columns.tolist()
concejos = [' '.join(element.split(' ')[1:]) for element in concejos]
```

```python
d = {}
for idx in df2.index[0:]:
    name, year = idx.rsplit(" ", 1)
    if name not in d:
        d[name] = df2.loc[[idx]]
        d[name].index = [year]
    else:
        temp_df = df2.loc[[idx]]
        temp_df.index = [year]
        d[name] = d[name].append(temp_df)
```

```python
nombredf = list(d.keys())
df_list = list(d.values())

for df in df_list:
    df.rename(columns={col: new_col for col, new_col in zip(df.columns, concejos)}, inplace=True)

results = {name: df for name, df in zip(nombredf, df_list)}

  
for name, df in results.items():
  df['Mancomunidad Cabo de Peñas'] = df['Carreño'] + df['Gozón']
  df['Mancomunidad Cinco Villas'] = df['Muros de Nalón'] + df['Pravia'] + df['Soto del Barco']
  df['Mancomunidad Comarca Avilés'] = df['Avilés'] + df['Castrillón'] + df['Corvera de Asturias'] + df['Illas']
  df['Mancomunidad Comarca de la Sidra'] = df['Bimenes'] + df['Cabranes'] + df['Colunga'] + df['Nava'] + df['Sariego'] + df['Villaviciosa'] 
  df['Mancomunidad de la Comarca del Sueve'] = df['Caravia'] + df['Parres'] + df['Piloña']
  df['Mancomunidad de la Comarca Vaqueira'] = df['Allande'] + df['Cudillero'] + df['Salas'] + df['Tineo'] + df['Valdés'] 
  df['Mancomunidad de los Concejos de cangas de onís, Amieva, Onís y Ponga'] = df['Amieva'] + df['Cangas de Onís'] + df['Onís'] + df['Ponga']
  df['Mancomunidad de los Concejos de Grado y Yermes y Tameza'] = df['Grado'] + df['Yernes y Tameza'] 
  df['Mancomunidad de Concejos de la Zona Suroriental de Asturias'] = df['Cabrales'] + df['Onís'] + df['Peñamellera Alta'] + df['Peñamellera Baja'] + df['Ribadedeva'] 
  df['Mancomunidad de los Concejos de Llanes y Ribadedeva'] = df['Llanes'] + df['Ribadedeva'] 
  df['Mancomunidad de los Concejos de Oscos-Eo'] = df['San Martín de Oscos'] +  df['San Tirso de Abres'] +  df['Santa Eulalia de Oscos'] +  df['Taramundi'] +  df['Vegadeo'] +  df['Villanueva de Oscos'] +  df['Castropol']
  df['Mancomunidad de los Concejos de Parres y Piloña'] = df['Parres'] + df['Piloña']
  df['Mancomunidad del Oriente de Asturias'] = df['Amieva'] + df['Cabrales'] + df['Cangas de Onís'] + df['Caravia'] + df['Llanes'] + df['Onís'] + df['Parres'] + df['Peñamellera Alta'] + df['Peñamellera Baja'] + df['Piloña'] + df['Ponga'] + df['Ribadedeva'] + df['Ribadesella']
  df['Mancomunidad de Montes de Peñamayor'] = df['Bimenes'] + df['Laviana'] + df['Nava'] + df['Piloña'] 
  df['Mancomunidad Occidental'] = df['Castropol'] + df['Coaña'] + df['Franco, El'] + df['Grandas de Salime'] + df['Illano'] + df['Navia'] + df['Pesoz'] + df['San Martín de Oscos'] + df['San Tirso de Abres'] + df['Santa Eulalia de Oscos'] + df['Tapia de Casariego'] + df['Taramundi'] + df['Valdés'] + df['Vegadeo'] + df['Villanueva de Oscos'] + df['Villayón']
  df['Mancomunidad de Servicios del Valle del Nalón'] = df['Caso'] + df['Langreo'] + df['Laviana'] + df['San Martín del Rey Aurelio'] + df['Sobrescobio'] 
  df['Mancomunidad Valle del Pigüeña'] = df['Belmonte de Miranda'] + df['Somiedo']
  df['Mancomunidad Valles del Oso'] = df['Proaza'] + df['Quirós'] + df['Santo Adriano'] + df['Teverga'] 
  # print("Name:", name)
  # print("Dataframe:")
  # display(df)
  

prueba = list(results.values())
results.keys()

dfTOTAL = prueba[0]
dfCastaño = prueba[1]
dfEucalipto = prueba[2]
dfPino_pinaster = prueba[3]
dfPino_radiata = prueba[4]
dfPino_silvestre = prueba[5]
dfQuercíneas = prueba[6]
dfOtras_especies = prueba[7]
```

```python
NOMBRE_CONCEJO = ['Villaviciosa', 'Mancomunidad Occidental']
fig = go.Figure()
for i in NOMBRE_CONCEJO:
  fig.add_trace(go.Bar(x=dfPino_radiata.index, y=dfPino_radiata[i].values, name=str('Pino_radiata: ')+dfPino_radiata[i].name))
  fig.add_trace(go.Bar(x=dfEucalipto.index, y=dfEucalipto[i].values, name=str('Eucalipto: ')+dfEucalipto[i].name))
fig.add_trace(go.Line(x=dfTOTAL.index, y=dfTOTAL['ASTURIAS'], name='Asturias', line={'color': 'darkblue', 'width': 1, 'dash': 'dash'}))
fig.update_layout(template='plotly_white', xaxis_title='Años', yaxis_title="Número de talas", title="Evolución en el número de talas en Asturias", barmode='stack')
fig.show()
```
